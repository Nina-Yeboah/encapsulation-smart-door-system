```python
class Staff:
    def __init__(self, s_name: str, access_code: str):
        self.s_name = s_name
        self.__access_code = access_code

    @property
    def access_code(self):
        """Allows authorized view of the access code."""
        authorized = input("Are you an authorized user? (yes/no): ").strip().lower()
        if authorized == "yes":
            return self.__access_code
        else:
            return "Access Denied: Unauthorized User."

    @access_code.setter
    def access_code(self, new_code: str):
        """Validates length before updating the private access code."""
        if len(new_code) == 6:
            self.__access_code = new_code
            print("Access code updated successfully.")
        else:
            print("Error: Invalid length! Access code must be exactly 6 characters.")
    def display_staff_info(self):
        """Displays public staff information without exposing sensitive data directly."""
        print("\n--- Staff Information ---")
        print(f"Staff Name: {self.s_name}")

        print("Access Code: [PROTECTED]")
        print("-------------------------")

if __name__ == "__main__":

    staff1 = Staff("Nina Yeboah", "A123B4")
    staff1.display_staff_info()

    print("\nAttempting to view access code...")
    print(f"Retrieved Code: {staff1.access_code}")

    print("\nAttempting to update code with an invalid length (123)...")
    staff1.access_code = "123"

    print("\nAttempting to update code with a valid length (XYZ987)...")
    staff1.access_code = "XYZ987"

    print("\nAttempting to view updated access code...")
    print(f"Retrieved Code: {staff1.access_code}")
```
